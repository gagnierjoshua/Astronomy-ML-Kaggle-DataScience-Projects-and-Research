```python
from astropy import units as u

import numpy as np

distances = np.array([4.0,32.5,44.7,1007.2])
print("distance array with no units",distances)

distancemeters=distances*u.m 
#variableaddingtounits*u.<unit you want>
print("array w/ units", distancemeters)

#if you have distances of travel
#and you know it took 30seconds to reach those
#distance, convert array into speed

speedthirty=distancemeters/(30*u.second)
#divided each column by 30 seconds, giving a speed
#in m/s 
print(speedthirty)

# .value to see just the numbers no unit
print("array value only:", speedthirty.value)

# . unit to see the units not the numbers
print("array unit only:", speedthirty.unit)

#how do you remove units or values? permanently 

speedthirty = speedthirty.value 
#to remove units, just redefine (make it equal to) just value of it
print("new:",speedthirty)
#you can do the same to remove numbers and just keep units with .unit
```

```python
#common astrony units 

#the nearest star (not sun) is 1.3 parsec away
alongdistance= 1*u.parsec
print("1 parsec:", alongdistance)

#to convert in astropy you use the code
#startingvariable.to(u.newunit)

alongdistance= alongdistance.to(u.km) 
#redefining/making my variable equal to the converted value
print("1 parsec in km:", alongdistance)

#an Au is the distance from the center of earth 
# to the center of the sun
alongdistance=alongdistance.to(u.au) 
print("1 parsec in astronmical units:", alongdistance)

#light year is the distance light travels going the speed of light
#for an entire year 
alongdistance=alongdistance.to(u.lyr)
print("1 parsec in lightyear:", alongdistance)
```

```python
#Astropy Constants (constant variables)

from astropy import constants as const 
#imports all of the astropy constants in one go, andto call them
#you just type const.unit name 

print(const.c) #print all the info of the constant 
print(const.c.value) #print/store(define) just value
print(const.c.unit) #print/store(define) just unit 
print(const.c.name) #print/store(define) just name 

print(const.GM_earth)
```

```python
#take speed of light and convert from m/s to km/s 
lightspeetconverted = const.c.to('km/s') 
print ("converted lightspeed:",lightspeetconverted)
```
